# Defining a Function in Python

```python
def simple():
    print("My first function")
```

```python
simple()
```

```python
def add_ten(x):
    result = x + 10
    return result
```

```python
add_ten(2)
```

```python
add_ten(5)
```

```python
def wage(w_hours):
    return w_hours * 25

def with_bonus(w_hours):
    return wage(w_hours) + 50
```

```python
wage(8), with_bonus(8)
```

```python
def plus_five(x):
    return x + 5
    
    
def m_by_3(x):
    return plus_five(x) * 3
    
plus_five(5), m_by_3(5)
```

```python
def add_10(m):
    if m >= 100:
        m = m + 10
        return m
    else:
        return "Save more!"
```

```python
add_10(100)
```

```python
add_10(22)
```

```python
def subtract_bc(a, b, c):
    result = a - b * c
    print("Parameter a equals", a)
    print("Parameter b equals", b)
    print("Parameter c equals", c)
    return result
```

```python
subtract_bc(10, 3, 2) # output: 4
```
