**Condition Statement, Functions, and Loops**

```python
def count(numbers):
    total = 0
    for n in numbers:
        if n < 20:
            total += 1
    return total
```

```python
nums = [1,35,12,24,31,51,70,100, 13, 17]
```

```python
count(nums)
```

**Iterating over Dictionaries**

```python
prices = {
    "box_of_spaghetti": 4,
    "lasagna": 5,
    "hamburger": 2
}

quantity = {
    "box_of_spaghetti": 6,
    "lasagna": 10,
    "hamburger": 0
}

money_spent = 0

for i in quantity:
    money_spent = money_spent + (prices[i] * quantity[i])


money_spent
```
