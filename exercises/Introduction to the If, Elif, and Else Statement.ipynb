**Else If, for Brief - Elif**

```python
def compare_to_five(y):
    if y > 5:
        return "Greater"
    if y < 0:
        return "Negative"
    elif y < 5:
        return "Less"
    else:
        return "Equal"
```

```python
print(compare_to_five(-3))
```

```python
print(compare_to_five(4))
```

```python
print(compare_to_five(20))
```

```python
print(compare_to_five(5))
```
