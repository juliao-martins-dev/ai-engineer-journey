```python
def raise_to_the_power_of_2(x):
    return x ** 2
```

```python
raise_to_the_power_of_2(3)
```

```python
raise_to_the_Power_of_2_lambda = lambda x: x ** 2
```

```python
raise_to_the_Power_of_2_lambda(3)
```

```python
(lambda x: x ** 2)(10)
```

```python
(lambda x: (2 + 5 * x ** 4) ** 2 / (x + 3) ** 3)(2)
```

```python
sum_xy = lambda x, y: x + y
```

```python
sum_xy(2, 3)
```

```python
sum_xy = lambda x, y: x + y(x)
```

```python
sum_xy(2, 3)
```
